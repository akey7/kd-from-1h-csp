# Kd from 1H CSP

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Concentration and CSP data

```python
csps = np.linspace(7., 8., 6)
ligand_mMs = [0., 50., 100., 200., 400., 800.]
```

## Binding isotherm plot

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 3.5))
ax.plot(ligand_mMs, csps)
ax.set_ylabel("CSP (ppm)", size=15)
ax.set_xlabel("ligand (mM)", size=15)
ax.set_title("Kd = 150 mM", size=20)
ax.axhline(7.5, xmax=150./800., color='r')
ax.axvline(150, ymax=.5, color='r')
ax.set_xlim(min(ligand_mMs), max(ligand_mMs))
ax.set_ylim(min(csps), max(csps))
```
